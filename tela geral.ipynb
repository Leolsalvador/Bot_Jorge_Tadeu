```python
from tkinter import*
from tkinter import messagebox
from openpyxl import Workbook, load_workbook
import win32com.client as win32

class Sistema():
    def __init__(self):
        self.lista_logins = ["adm"]
        self.lista_senhas = ["adm"]
        #Criação da função

        self.janela = Tk()
        self.janela.geometry("200x300")
        self.janela.title('Monitoramento')

        #Título da Janela

        label_descricao = Label(self.janela, text="Área de Login")
        label_descricao.pack()
        label_descricao.grid(row=1, column=1,padx = 50, pady=10, sticky='nswe', columnspan =4  )

        label_login = Label(self.janela,text="Digite o seu login")
        label_login.grid(row=2, column=1,padx = 50, pady=10, sticky='nswe', columnspan =4  )

        self.entry_login = Entry(self.janela,)
        self.entry_login.grid(row=3,column=2, padx=40, pady=10, sticky='nswe', columnspan = 4)

        label_senha = Label(self.janela,text="Digite sua senha")
        label_senha.grid(row=4, column=1,padx = 50, pady=10, sticky='nswe', columnspan =4  )

        self.entry_senha = Entry(self.janela,)
        self.entry_senha.grid(row=5,column=2, padx=40, pady=10, sticky='nswe', columnspan = 4)

        botao_login =Button(self.janela,text="Fazer login", command=self.login)
        botao_login.grid(row=6,column=2,padx = 10, pady=10,sticky='nswe', columnspan =4)

        botao_cad = Button(self.janela,text="Fazer cadastro", command=self.cadastro)
        botao_cad.grid(row=7,column=2,padx = 10, pady=10,sticky='nswe', columnspan =4)
        
        self.janela.mainloop() 
        
    def login(self):
        login = self.entry_login.get()
        senha = self.entry_senha.get()
  
        wb = load_workbook(r"Login.xlsx")
        lista_logins = []
        lista_senhas = []

        ws = wb ['Logins'] 
        ws = wb.active
        maxrow = ws.max_row
        if login == adm and senha == adm:
            self.adm()
        
        for i in range(1, maxrow + 1): 
            col_nome = ws.cell(row = i, column = 1) 
            lista_logins.append(col_nome.value)

        for i in range(1, maxrow + 1): 
            col_senha = ws.cell(row = i, column = 2) 
            lista_senhas.append(col_senha.value)

        usuario_encontrado = True
        for i in lista_logins: 
            if i == login:
                for j in lista_senhas:
                    if j == senha:
                        self.chatbot()
                    else:
                        usuario_encontrado = False
            else:
                usuario_encontrado = False
        
        if usuario_encontrado == False:
            messagebox.showinfo(title='Login não encontrado',message='Verifique se o login e senha estão corretos')

    def cadastro(self):
        self.cadastro = Toplevel()
        self.cadastro.geometry("200x300")
        self.cadastro.title('Cadastro')
        
        label_cad = Label(self.cadastro,text="Digite o usuário")
        label_cad.pack()
        label_cad.grid(row=2, column=1,padx = 50, pady=10, sticky='nswe', columnspan =4  )
        
        self.entry_usuario = Entry(self.cadastro,)
        self.entry_usuario.grid(row=3,column=2, padx=40, pady=10, sticky='nswe', columnspan = 4)

        label_senhaC = Label(self.cadastro, text="Digite sua senha")
        label_senhaC.grid(row=4, column=1,padx = 50, pady=10, sticky='nswe', columnspan =4  )

        self.entry_senhaC = Entry(self.cadastro)
        self.entry_senhaC.grid(row=5,column=2, padx=40, pady=10, sticky='nswe', columnspan = 4)

        botao_cad = Button(self.cadastro, text="Cadastrar", command=self.cad)
        botao_cad.grid(row=7,column=2,padx = 10, pady=10,sticky='nswe', columnspan =4)
        
        

    def cad(self): 
        wb = load_workbook(r"Login.xlsx")
        lista_logins = []
        lista_senhas = []

        ws = wb ['Logins'] 
        ws = wb.active
        maxrow = ws.max_row
        
        for i in range(1, maxrow + 1): 
            col_nome = ws.cell(row = i, column = 1) 
            lista_logins.append(col_nome.value)

        for i in range(1, maxrow + 1): 
            col_senha = ws.cell(row = i, column = 2) 
            lista_senhas.append(col_senha.value)
        
        global usuario, senhaC
        usuario = self.entry_usuario.get()
        senhaC = self.entry_senhaC.get()
        print("Usuario: ",usuario)
        print("Senha Cadastro: ",senhaC)
        self.lista_logins.append(usuario)
        self.lista_senhas.append(senhaC)
        to_append = [f'{usuario}',f'{senhaC}']
        ws.append(to_append) 
        my_value = ws[f'A{maxrow}'].value
        my_value1 = ws[f'B{maxrow}'].value  
        wb.save(r'Login.xlsx')
        messagebox.showinfo(title='Cadastro bem sucedido',message='Seu perfil foi cadastrado com sucesso')
        self.chatbot()

    def chatbot(self):
        self.chatbot = Tk()
        w, h = self.chatbot.winfo_screenwidth(), self.chatbot.winfo_screenheight()
        self.chatbot.geometry("%dx%d+0+0" % (w, h))
        self.chatbot.title('Chatbot')  
        
        label_des = Label(self.chatbot, text = "Olá eu sou o Jorge Tadeu, seu bot de erros. Esse sistema consiste em mostrar\n todos os erros apresentados em alguns sistemas do INSS. No momento estou testeando os seguintes sistemas:")
        label_des.grid(row=0, column=5)
        label_S1 = Label(self.chatbot, text = "O Chatbot Hello")
        label_S1.grid(row=2, column=5)
        label_S2 = Label(self.chatbot, text = "O Meu INSS")
        label_S2.grid(row=3, column=5)
        label_S3 = Label(self.chatbot, text = "O SAG Internet")
        label_S3.grid(row=4, column=5)
        label_S4 = Label(self.chatbot, text = "O SAT Central")
        label_S4.grid(row=5, column=5)
        label_S5 = Label(self.chatbot, text = "O Novo SEC")
        label_S5.grid(row=6, column=5)
        
    def adm(self):
        self.adm = Tk()
        self.chatbot.geometry("%dx%d+0+0" % (w, h))
        self.adm.title("Área do Administrador")
        
        
app = Sistema()
```
